```python
# Importing the dataset and display the first 10 rows
import pandas as pd
graffiti = pd.read_csv('DSNY_Graffiti_Tracking.csv')
graffiti.head(10)
```

```python
# Displaying the number of observations in the graffiti dataset
len(graffiti)
```

```python
# Getting general information about the graffiti dataset
graffiti.info()
```

```python
# Getting the number of NULL values in the graffiti dataset
graffiti.isnull().sum()
```

As we can see, not even half of the observations in the dataset have complete data. We will extract the observations with complete data and see what that leaves us with.

```python
# Dropping any rows with a NA in the dataset to ensure completeness of the data
graffiti_complete = graffiti.dropna()
graffiti_complete.head(10)
```

```python
# Displaying the new number of observations in the dataset
len(graffiti_complete)
```

```python
# Getting general information about the dataset
graffiti_complete.info()
```

```python
# Calculating the number of NULL values in the dataset
graffiti_complete.isnull().sum()
```

As we can see, we have dropped all observations with NAs in them. Thus, we are left with only the observations with complete data.
